```python
import pandas as pd
import numpy as np
import joblib
```

```python
df=pd.read_csv(r'mushroom_csv.csv')
df.info()
```

```python
df.drop(columns=['stalk-root','population','ring-type'],inplace=True)
df.rename(columns={'bruises%3F':'bruises'},inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_enc=pd.DataFrame()
for col in df.columns:
    df_enc[col]=le.fit_transform(df[col])
    
df_enc
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
df_enc=mms.fit_transform(df_enc)
df_enc=pd.DataFrame(df_enc,columns=df.columns)
```

```python
y=df_enc.pop('class')
X=df_enc

from sklearn.model_selection import train_test_split
X_tr,X_ts,y_tr,y_ts=train_test_split(X,y,test_size=0.30,random_state=66)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import KFold


cv = KFold(n_splits=5, shuffle=True, random_state=42)

pipeline = Pipeline([
    ('feature_selection', SelectKBest(score_func=chi2)),
    ('model', RandomForestClassifier())
])

param_grid = {
    'feature_selection__k': [5, 10, 15],
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [5, 10, 20, None],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=cv, n_jobs=-1)
grid_search.fit(X_tr, y_tr)

print("Best hyperparameters: ", grid_search.best_params_)
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))

mask = grid_search.best_estimator_.named_steps['feature_selection'].get_support()
selected_columns = X.columns[mask]
print("Selected feature columns: ", selected_columns)
```

```python

df_sel=df[selected_columns]
df_sel['class']=df['class']
```

```python
df_sel
```

### Final Model Build

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelEncoder
le_dict = {col: LabelEncoder() for col in selected_columns}
for col in selected_columns:
    le_dict[col].fit_transform(df[col])

df_nd=pd.DataFrame()
for coll in selected_columns:
    df_nd[coll]=le_dict[coll].transform(df[coll])
```

```python
from sklearn.preprocessing import MinMaxScaler
mms_nd=MinMaxScaler()
df_nd=mms_nd.fit_transform(df_nd)
df_nd=pd.DataFrame(df_nd,columns=selected_columns)
```

```python
model_nd=RandomForestClassifier(max_depth=None,min_samples_leaf=1,min_samples_split=5,n_estimators=200)
model_nd.fit(df_nd,y)
```

```python
from pickle import dump
dump(le_dict, open('encoder.pkl', 'wb'))
dump(mms_nd,open('scaler.pkl','wb'))
dump(model_nd,open('model.pkl','wb'))
```
